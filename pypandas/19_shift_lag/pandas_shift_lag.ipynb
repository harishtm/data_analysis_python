```python
import pandas as pd
```

```python
dfdata = {
    'Date': ['2017-08-15', '2017-08-16', '2017-08-17', '2017-08-18', '2017-08-21', '2017-08-22',
             '2017-08-23', '2017-08-24', '2017-08-25', '2017-08-28'],
    'Price': ['171.00', '170.00', '166.91', '167.41', '167.78', '169.64', '168.71', '167.74', '166.32', '167.24']
}
```

```python
df = pd.DataFrame(dfdata)
```

```python
df
```

```python
# Shift specific column
df.Price = df.Price.shift(1)
```

```python
df
```

```python
# Will shift all columns
df.shift(1)
```

```python
df['Previous Day Price'] = df['Price'].shift(1)
```

```python
df
```

```python
df['Price'] = df['Price'].fillna(0)
df['Previous Day Price'] = df['Previous Day Price'].fillna(0)
```

```python
df['Price Change'] = df['Price'].astype(float) - df['Previous Day Price'].astype(float)
```

```python
df
```

```python
df['5 days return'] = (df['Price'].astype(float) - df['Price'].shift(5).astype(float)) * 100 / df['Price'].shift(5).astype(float)
```
