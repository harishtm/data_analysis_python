```python
import pandas as pd
```

```python
data = pd.read_csv('employees.csv')
```

```python
data.columns =[column.replace(" ", "_") for column in data.columns]
```

```python
data['Start_Date'] = pd.to_datetime(data['Start_Date'], format="%m/%d/%Y")
```

```python
# filtering with query method
data.query('Senior_Management == True')
```

```python
# filtering with query method with multiple conditions
data.query('Senior_Management == True \
           and Gender == "Male" \
           and Team == "Marketing" \
           and Salary > 90000')
```

```python
# Date Based query start_date > 2000
data.query('Senior_Management == True \
           and Gender == "Male" \
           and Team == "Marketing" \
           and Salary > 90000 \
           and Start_Date > 2000')
```
