```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
df.day[0]
```

```python
# Checking the type of column
type(df.day[0])
```

```python
# Setting a column as a index
df.set_index('day', inplace=True)
```

```python
df
```

```python
# Fill na with 0
new_df = df.fillna(0)
new_df
```

```python
# Filling 0 to an event is not correct as event is not an integer
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
})
new_df
```

```python
# Filling 0 mignt not be a better solution to get the accurate mean / median values (ForwardFill)
new_df = df.fillna(method='ffill')
new_df
```

```python
# Similarly based on the requirement backward fill can also be done
new_df = df.fillna(method='bfill')
new_df
```

```python
# NaN values can be filled with axis as well (i.e) the corresponding column values
new_df = df.fillna(method='bfill', axis='columns') # axis is either index or columns
new_df
```

```python
# During ffill or bfill the forward or backward values can be limited to specific number of cells
new_df = df.fillna(method='ffill', limit=1)
new_df
```

```python
# Interpolate gives the average between the two cells
new_df = df.interpolate()
new_df
```

```python
# Different methods associated with interpolate are
# {‘linear’, ‘time’, ‘index’, ‘values’, ‘nearest’, ‘zero’,‘slinear’, ‘quadratic’, ‘cubic’, ‘barycentric’, ‘krogh’, ‘polynomial’, ‘spline’ ‘piecewise_polynomial’, ‘pchip’}
new_df = df.interpolate(method='time')
new_df # Temperature calculated based on the days(2017-01-01 to 2017-01-04)
```

__Notice that in above temperature on 2017-01-04 is 29 instead of 30 (in plain linear interpolate).There are many other methods for interpolation such as quadratic, piecewise_polynomial, cubic etc. refer "dataframe interpolate" to see complete documentation__

```python
# drop na
new_df = df.dropna()
new_df
```

```python
# If all values are NA, drop that row or column.
# how : {‘any’, ‘all’}, default ‘any’
new_df = df.dropna(how='all')
new_df
```

```python
# drops rows which has more than 2 NaN column values
new_df = df.dropna(thresh=2)
new_df
```
