```python
import pandas as pd
```

```python
df = pd.read_csv("weather_by_cities.csv")
df
```

__For the above dataset__

__1. What was the maximum temperature in each of these 4 cities?__<br>
__2. What was the average windspeed in each of these 4 cities?__

```python
g = df.groupby('city')
g
```

```python
for city, data in g:
    print("\n")
    print("city: ", city)
    print("data: ", data)
```
