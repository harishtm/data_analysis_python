__Matplotlib is a library for 2-dimensional plotting with Python__

```python
x = [1,2,3,4,5,6,7,8]
y = [50,51,52,63,64,65, 70,75]
```

__Parametrs:__

x, y  - array like or scalar

fmt - str, optional (A format string e.g 'ro' for red circle, 'gs' for green square and so on reffer color, marker, linestyle for more options)

data - indexable object, optional An object with lablled data

__Graph with x and y axis __

```python
plt.plot(x,y)
```

```python
# Setting lables to x axes, y axes and title to graph with plot color, line width, line style
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x, y, color='green', linewidth=5, linestyle='dotted')
```

```python
# The default format string gives us a solid blue line
plt.plot([2, 3, 4, 5])
plt.xlabel('Actual birth weight')
plt.ylabel('Estimated birth weight')
plt.show()
```

__The third argument will let you choose the color and the line type of the plot ‘gs’ here gives us green squares__

```python
plt.plot([2,3,4,5],[3,8,10,12],'gs')
```

__Similarly, we can plot green, red, and blue orbs, squares, and triangles too lets take another example__

```python
t = np.arange(0, 5, 0.2) # 0 to 5 with 0.2 increment
t
```

```python
plt.plot(t, t, 'r--') # -- indicates dashed line style
```

```python
plt.plot(t, t, 'r--', t, t**3, 'b^', t, t**2, 'gs')
```

__Ploting 3 different lines on a graph__

The combinations are 'Colors', 'Markers' and 'line style'

r-- : indicates (red color, -- dashed line style)

b^ : indicates (blue color, ^ triangle_up marker)

gs : indicates (green color, s square marker)

__Using a data keyword argument, we can produce Python plots with strings that relate to some variables. We can access these variables with strings__

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
```

```python
data['b'] = data['a']+10*np.random.randn(50)
data['d']=np.abs(data['d'])*100
```

```python
data
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data) # scatter plot
plt.show()
```

__LineWidth__

```python
plt.plot([1, 2, 3], [2, 4, 9], linewidth=8.0)
plt.plot([1, 6, 7], [2, 8, 9], linewidth=4.0)
```

__Alpha - We can use the alpha channel to create prettier plots by softening colors.__

```python
plt.plot([1,2,3],[2,4,9],alpha=5.5)
```

__Antialiased If you’ll look closely, the lines look quite smooth. But we can turn antialiasing off- this will show us the aliasing in the lines.__

```python
plt.plot([1,2,3],[2,4,9],antialiased=True)
```

__Color You can set the color of the plot with this parameter.__

```python
plt.plot([1,2,3],[2,4,9],color='Chartreuse')
plt.plot([1,5,6],[2,1,7],color='purple')
plt.plot([5,4,3],[2,1,6],color='orange')
```

__Dashes__

```python
plt.plot([1,2,3],[2,4,9],dashes=[1,2,4,4])
plt.show()
```

__Linestyle You can choose the style of line you want for your plot.__

```python
plt.plot([1,2,3],[2,4,9],linestyle='steps')
plt.show()
```

```python
plt.plot([1,2,3],[2,4,9],linestyle=':')
plt.show()
```

__Marker - A marker will let you select what symbol you want to display at breakpoints and bends__

```python
plt.plot([1,2,3],[2,4,9],marker='+')
```

__Markeredgecolor - You can also decide on a color for your markers’ edges__

```python
plt.plot([1,2,3],[2,4,9],marker='+',markeredgecolor='brown')
```

__Markeredgewidth - This lets us decide how thick we want our markers to be.__

```python
plt.plot([1,2,3],[2,4,9],marker='+',markeredgewidth=5.5)
plt.show()
```

__Markerfacecolor and Markersize - This will let you choose which color to fill in your marker and what size to keep it.__

```python
plt.plot([1,2,3],[2,4,9],marker='.',markerfacecolor='orange',markersize=23.0)
plt.show()
```

__Markevery - This parameter lets us decide at what parameters we want to put markers.__

```python
plt.plot([1,2,3],[2,4,9],marker='.',markerfacecolor='orange',markersize=13.0,markevery=2)
plt.show()
```

__Zorder - This lets us decide which plots will show up in the front and which to send to back.__

```python
plt.plot([1,2,3],[2,4,9],zorder=2,linewidth=4)
plt.plot([1,2,6,9],[2,4,9,10],zorder=1,linewidth=4)
plt.show()
```

```python
plt.plot([1,2,3],[2,4,9],zorder=1,linewidth=4)
plt.plot([1,2,6,9],[2,4,9,10],zorder=2,linewidth=4)
plt.show()
```

In this graph, in the area enclosed by the points (1,2) and (2,4), both plots cover the same points, but the one in orange is in the front. This is because we give it a higher zorder. This is like z-index in CSS.
