```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, color='red')
plt.show()
```

```python
plt.plot(x, y, color='red', linestyle='--')
plt.show()
```

```python
plt.plot(x, y, color='red', linestyle='') # as the line style is empty the graph is not visible
plt.show()
```

```python
# line style is empty but still the marker is provided hence graph is visible
plt.plot(x, y, color='red', linestyle='', marker='+')
plt.show()
```

```python
plt.plot(x, y, color='red', linestyle='', marker='+', markersize=20)
plt.show()
```

```python
# g< - (i,e) g - green and < - triangle_left marker

plt.plot(x, y, 'g<', alpha=0.5) # alpha can be specified on a scale 0 to 1
```
