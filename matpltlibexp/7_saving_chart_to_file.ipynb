```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_values = [1400, 600, 300, 410, 250]
exp_lables = ['Home Rent', 'Food', 'Phone Bills', 'Car', 'Miscellaneous']
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_lables, radius=1.5, autopct='%0.1f%%', explode=[0,0,0,1,2])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_lables, radius=1.5, autopct='%0.1f%%', explode=[0,0,0,1,2])
plt.savefig('piechart.jpg', bbox_inches='tight', pad_inches=1, transparent=True)
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_lables, radius=0.75, autopct='%0.2f%%')
plt.show()
```

__You can also save in pdf format and to a different location on your computer by specifying absolute path__

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_lables, radius=2, autopct='%0.2f%%', explode=[0,0,0,0.1,0.2])
plt.savefig('/home/user/Desktop/piechart.pdf', bbox_inches="tight", pad_inches=10, transparent=True)
```
