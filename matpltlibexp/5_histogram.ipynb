```python
%matplotlib inline
import matplotlib.pyplot as plt
```

plt.hist(x, bins=None, range=None, density=None, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, normed=None, hold=None, data=None, **kwargs)

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8) # by default number of bins is set to 10
```

__bins parameter__ - bins : integer or sequence or 'auto', optional

```python
plt.hist(blood_sugar, rwidth=0.8, bins=20)
```

```python
plt.hist(blood_sugar, rwidth=0.5, bins=4)
```

__Histogram showing normal, pre-diabetic and diabetic patients distribution__

 - 80-100: Normal
 - 100-125: Pre-diabetic
 - 80-100: Diabetic

```python
plt.xlabel('Sugar Level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')

plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.95, color='g')
```

__Multiple data samples in a histogram__

```python
plt.xlabel('Sugar Level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95,
         color=['green', 'orange'], label=['men', 'women'])
```

```python
plt.xlabel('Sugar Level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95,
         color=['green', 'orange'], label=['men', 'women'])
plt.legend() # draws the label
```

__histtype=step__

```python
plt.xlabel('Sugar Level')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Chart')

plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.95, histtype='step')
```

__horizontal orientation__

```python
plt.xlabel('Number of Patients')
plt.ylabel('Sugar Level')
plt.title('Blood Sugar Chart')

plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.95, orientation='horizontal')
```
