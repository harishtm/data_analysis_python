```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

__Simple bar chart showing companies revenue__

```python
company = ['Amazon', 'FlipKart', 'SnapDeail', 'Myntra']
revenue = [150, 125, 85, 60]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos, revenue, label='Revenue in Cr.')

plt.xticks(xpos, company)
plt.ylabel('Revenue in (Cr)')
plt.title('Sales for year 2017-2018')
plt.legend()
plt.show()
```

__Multiple bars showing revenue and profit__

```python
profit = [40, 35, 20, 10]
plt.bar(xpos-0.2, revenue, width=0.4, label='Revenue')
plt.bar(xpos+0.2, profit, width=0.4, label='Profit')
plt.xticks(xpos, company)
plt.ylabel('Revenue in (Cr.)')
plt.title('Sales Graph with Profit')
plt.legend()
```

__Horizontal bar__

```python
plt.barh(xpos, revenue, label='Revenue')
plt.yticks(xpos, company)
plt.ylabel('Revenue in (Cr.)')
plt.title('Sales Data')
plt.legend()
```
