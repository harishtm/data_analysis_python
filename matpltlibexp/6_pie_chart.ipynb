```python
%matplotlib inline
import matplotlib.pyplot as plt
```

__Draw pie chart to track down home expenses__

```python
exp_values = [1400, 600, 300, 410, 250]
exp_lables = ['Home Rent', 'Food', 'Phone Bills', 'Car', 'Miscellaneous']
plt.pie(exp_values, labels=exp_lables)
plt.show()
```

__Draw pie chart as perfect circle__

```python
plt.pie(exp_values, labels=exp_lables, shadow=True)
plt.axis('equal') # to draw the exact circle
plt.show()
```

__Show percentage for every pie. specify the radius to increase chart size__

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_lables, shadow=True, autopct='%1.1f%%', radius=1.5)
plt.show()
```

__Explode__

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_lables, autopct='%1.2f%%', radius=1.5, explode=[0, 0, 0, 0.2, 0.3])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_lables, autopct='%1.2f%%', radius=1.5, explode=[0, 0.5, 0.1, 0.2, 0.3])
plt.show()
```

__counterclock and angle properties__

```python
plt.axis('equal')
plt.pie(exp_values, labels=exp_lables, radius=1.25, autopct='%1.3f%%', counterclock=True,
        startangle=45, explode=[0, 0.5, 0.1, 0.2, 0.3])
plt.show()
```
