__Using RandomForest  classifier for prediction__

```python
df_final = df.drop(['salary'], axis=1).copy()
```

```python
df_final.sales.value_counts()
```

```python
pd.get_dummies(df_final, columns=['sales']).head()
```

```python
df_final = pd.get_dummies(df_final, columns=['sales'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
y = df_final['left']
X = df_final.drop(['left'], axis=1)
```

__Takes 67% data as training data and 33% data as test data which is specified as test_size__

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
```

```python
X_train = X_train.drop(['colors'], axis=1)
```

```python
clf.fit(X_train, y_train)
```

```python
X_test = X_test.drop(['colors'], axis=1)
```

```python
y_pred = clf.predict(X_test)
```

```python
my_list = list(zip(list(y_test),list(y_pred)))
```

```python
accuracy_score(y_test, y_pred)
```

```python
features = df_final.columns
importance = clf.feature_importances_
```

```python
indices = np.argsort(importance)
```

```python
plt.title('Feature Importance')
plt.barh(range(len(indices)), importance[indices], color='b', align='center')
plt.yticks(range(len(indices)), features)
plt.xlabel('Relative Importance')
plt.show()
```
